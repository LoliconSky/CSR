```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 从 1 到 10 生成 100 个数（线性增长的）
x = np.linspace(1, 10, 100)
y = 2*x + 1

# 绘制
plt.plot(x, y)
plt.show()
```

```python
x2 = np.linspace(-1, 1, 100)
y2 = x2**2

# 新图表，否则 plot 画的会在一张表上
plt.figure(figsize=(4,4))

plt.xlabel('I am X')
plt.ylabel('I am Y')
plt.plot(x2, y2, color='red', linewidth=1.0, linestyle='--')  # - 实线，-- 虚线
plt.show()
```

```python
for x in np.linspace(-1, 1, 10):
    print(x)
```

```python
x2 = np.linspace(-3, 3, 100)
y2 = x2**2

plt.xlabel('I am X')
plt.ylabel('I am Y')

# 限制范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

plt.plot(x2, y2)

# 设置 x y 轴坐标
plt.xticks(np.linspace(-2, 2, 11))
plt.yticks([-1, 0, 1, 2], ['aa', 'bb', 'cc', 'dd'])

# 设置边框
ax = plt.gca()
ax.spines['right'].set_color('red')
ax.spines['top'].set_color('none')

plt.show()
```

```python
x2 = np.linspace(-3, 3, 100)
y2 = x2**2

# 限制范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

plt.plot(x2, y2)

# 设置边框
ax = plt.gca()
# 设置 x 轴位置
ax.xaxis.set_ticks_position('bottom')

# 设置轴的相对位置（原点位置）
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

plt.show()
```

```python
# 从 1 到 10 生成 100 个数（线性增长的）
x = np.linspace(0, 10, 100)
y = 2*x + 1

plt.xlim((0, 10))
plt.ylim((0,25))

# 绘制
pl, = plt.plot(x, y)

# 设置图例，位置自动
plt.legend(handles=[pl,], labels=['desc',], loc='best')

# 画点
x0 = 4.5
y0 = 2*x0 + 1
plt.scatter(x0, y0, s=50, color='b')  # blue  s=size
# 画虚线 (--)，黑色 (k)，宽度 2
plt.plot([x0, x0], [y0, 0], 'k--', lw=2)

plt.show()
```

```python
# 散点图
plt.scatter(np.arange(30), np.random.normal(1, 30, 30), s=40, c='b', alpha=0.5)

# 取消坐标显示
plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# 直方图
x = np.linspace(1, 10, 5)
y = 2*x + 1

# 绘制
plt.bar(x, y)
# zip 聚合 x 和 y
for x,y in zip(x, y):
    plt.text(x, y, '%d' %y, ha='center', va='bottom')

plt.figure()
x = np.linspace(1, 10, 5)
y = 2*x + 1
plt.bar(x, -y, facecolor='#ccc')

plt.show()
```
